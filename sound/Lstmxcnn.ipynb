```python
train = pd.read_csv('spec_df.csv')
```

```python
train_image = []
for i in tqdm(range(train.shape[0])):
    img = image.load_img(train['iname'][i], target_size=(28, 28, 3), grayscale=False)
    img = image.img_to_array(img)
    img = img/255
    train_image.append(img)
print(train_image)
X = np.array(train_image)
```

```python
len(train_image[0][0][0])
```

```python
le = LabelEncoder() 
  
train['label'] = le.fit_transform(train['label']) 

y=train['label'].values
y = to_categorical(y)
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=50, test_size=0.2, shuffle = True)
```

```python
X_train.shape
```

```python
from tensorflow.keras.layers import TimeDistributed
from tensorflow.keras.layers import AveragePooling1D
from tensorflow.keras.losses import categorical_crossentropy
from tensorflow.keras.optimizers import RMSprop
```

```python
x_train = np.array(X_train).reshape((368, 28, -1))
```

```python
x_train.shape
```

```python
batch_size = 128
nb_epoch = 100

# Parameters for MNIST dataset
img_rows, img_cols = 28, 28
nb_classes = 3

# Parameters for LSTM network
nb_lstm_outputs = 30
nb_time_steps = img_rows
dim_input_vector = img_cols
```

```python

input_shape = (nb_time_steps, dim_input_vector)
input_shape
```

```python
print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
x_train = np.array(X_train).reshape((368, 28, -1))
x_test = np.array(X_test).reshape((93, 28, -1))
```

```python
from tensorflow.keras.layers import TimeDistributed
from tensorflow.keras.layers import AveragePooling1D
from tensorflow.keras.losses import categorical_crossentropy
from tensorflow.keras.optimizers import RMSprop
```

```python
model = Sequential()
model.add(LSTM(64, input_shape=(28, 84), return_sequences=True, implementation=2))
model.add(TimeDistributed(Dense(1)))
model.add(AveragePooling1D())

model.add(Flatten())

model.add(Dense(3, activation='softmax'))
model.compile(loss=categorical_crossentropy, optimizer=RMSprop(lr=.01),metrics=['accuracy'])
```

```python
# Train
history = model.fit(x_train, y_train, epochs=50, batch_size=batch_size, shuffle=True, verbose=1, validation_data=(x_test, y_test))

# EvaluateX_train, y_train, epochs=35, validation_data=(X_test, y_test)
evaluation = model.evaluate(x_test, y_test, batch_size=batch_size, verbose=1)
print('Summary: Loss over the test dataset: %.2f, Accuracy: %.2f' % (evaluation[0], evaluation[1]))
```

```python
scores = model.evaluate(x_test, y_test, verbose = 0)
print("Model evaluation accuracy: ", round(scores[1]*100))
scores = model.evaluate(x_train, y_train, verbose = 0)
print("Model training accuracy: ", round(scores[1]*100))
```
